```python
import numpy as np
import pandas as pd
import pickle

from glob import glob
from PIL import Image
```

```python
saved_model = pickle.load(open('plant_disease_model.pickle', 'rb'))
```

```python
clf = saved_model
```

```python
img_path = glob('dataset/healthy/*')[0]
incoming_data = Image.open(img_path)

incoming_data
```

```python
incoming_data = incoming_data.resize([32, 32])
incoming_data = np.array(incoming_data).reshape(1, -1)

incoming_data
```

# Prediction

```python
predicted_target = clf.predict(incoming_data)
```
