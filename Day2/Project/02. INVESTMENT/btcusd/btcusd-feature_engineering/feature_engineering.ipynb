```python
import pandas as pd
import numpy as np
import talib
import warnings
```

```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('ราคาย้อนหลัง BTC_USD Bitfinex.csv', thousands=',')
df.columns = ['Date', 'Close', 'Open', 'High', 'Low', 'Volume', '%Change']

data = df[['Date', 'Open', 'High', 'Low', 'Close']]
data_lag1 = data.shift(1)

data['Open_0'] = data['Open']
data['STD5_Open_0'] = talib.STDDEV(data['Open'], timeperiod=5)
data['RSI14_Open_0'] = talib.RSI(data['Open'], timeperiod=14)
data['ATR14_0'] = talib.ATR(data_lag1['High'], data_lag1['Low'], data_lag1['Close'], timeperiod=14)

data['High_0'] = data['High']
data['Low_0'] = data['Low']

data_out = data.copy()
data_out = data_out.dropna()
data_out = data_out.drop(['Open', 'High', 'Low', 'Close'], axis=1)
data_out.to_csv('btcusd_dataset.csv', index=False)
data_out.to_csv('../btcusd_dataset.csv', index=False)

data_out
```
