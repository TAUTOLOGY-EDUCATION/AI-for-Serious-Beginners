# 3. feature extraction

## 3.1. define label

```python
yes_label = ["VeryMildDemented", "MildDemented", "ModerateDemented"]
no_label = ["NonDemented"]
```

```python
image_width, image_height = img.size
```

## 3.2. check data from csv

```python
all_data_path = [os.path.normpath(i) for i in glob("../dataset/*")]
is_has_data = f"..\\dataset\\data.csv" in all_data_path
```

```python
is_has_data
```

## 3.3. read and transform

```python
if not is_has_data:
    X = np.empty((0, image_width * image_height))
    y = np.empty((0, 1))

    for _class in no_label:
        folder_path = [
            os.path.normpath(i) for i in glob("../dataset/train/" + _class + "/*")
        ]
        print(f"load class : {_class}")
        for file_path in tqdm(folder_path):
            img = Image.open(file_path)
            array = np.array(img.getdata())
            X = np.vstack((X, array))
            y = np.vstack((y, "no"))

    for _class in yes_label:
        folder_path = [
            os.path.normpath(i) for i in glob("../dataset/train/" + _class + "/*")
        ]
        print(f"load class : {_class}")
        for file_path in tqdm(folder_path):
            img = Image.open(file_path)
            array = np.array(img.getdata())
            X = np.vstack((X, array))
            y = np.vstack((y, "yes"))
```

## 3.4. save dataframe

```python
if not is_has_data:
    data = pd.DataFrame(
        X, columns=[f"pixel_{i}" for i in range(image_width * image_height)]
    )
    data["label"] = y
    data.to_csv(f"..\\dataset\\data.csv", index=False)
```
